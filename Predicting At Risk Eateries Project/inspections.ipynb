```python
import re
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import time
```

```python
url_domain = 'https://healthspace.com/'
```

```python
url_main = 'https://healthspace.com/Clients/Ohio/Butler/Web.nsf/Food-CityList?OpenView&Count=999&'
```

```python
location_links = []
```

```python
r_main = requests.get(url_main)
soup = BeautifulSoup(r_main.text, 'html.parser')
all_a = soup.find_all('a')
for a in all_a:
    location_links.append(''.join([url_domain, a['href']]))
```

```python
location_link = [location + '&start={}' for location in location_links[3:]]
location_link
```

```python
location_link[0].format(1)
```

```python
list_links = []
```

```python
for location in location_link:
    print(location)
    for i in range(1, 10000, 50):
        r_location = requests.get(location.format(i))
        if r_location.text.find('No documents found') != -1:
            break
        soup = BeautifulSoup(r_location.text, 'html.parser')
        tds = soup.find_all('td', {'class': 'col-xs-4 col-sm-4 col-md-4'})
        for td in tds:
            a = td.find('a')
            list_links.append(''.join([url_domain, a['href']]))
        # time.sleep(0.5)
```

## Assemble the Dataframe

```python
## initiate the list
name_for_append = []
inspection_type = []
inspection_date = []
inspection_count = []
```

```python
# list_links_test[0]
```

```python
for l in list_links: 
    r = requests.get(l)
    soup = BeautifulSoup(r.text, 'html.parser')
    print(l)
    inspections = soup.find_all('td', {'class': 'col-xs-4 col-sm-4 col-md-4'})
    for inspection in inspections: 
        inspection_details = inspection.find('a')
        inspection_type.append(inspection_details.contents[0])
        
    inspection_date_count = soup.find_all("td", {"valign" : "top"})
    for index,i in enumerate(inspection_date_count):
        if index % 2 == 0:
            inspection_date.append(i.contents[0][2:])
        else:
            inspection_count.append(i.contents[0][2:])
            
    no_to_append = int(len(inspection_date_count)/2)
    details_html = soup.find_all('td', {'class': 'col-xs-8 col-sm-8 col-md-9'})
    details = [details_html[i].contents[0] for i in range(len(details_html))]
    details[0]
    for i in range(0,no_to_append):
        name_for_append.append(details[0])
```

```python
#name_for_append,inspection_type,inspection_date ,inspection_count
```

```python
inspections_pd = pd.DataFrame(columns=["Facility Name","Inspection Type","Inspection Date","Inspection Count"])
```

```python
inspections_pd["Facility Name"] = name_for_append
inspections_pd["Inspection Type"] = inspection_type
inspections_pd["Inspection Date"] = inspection_date
inspections_pd["Inspection Count"] = inspection_count
```

```python
inspections_pd
```

```python
df_all = facility_info.merge(inspections_pd.drop_duplicates(), on='Facility Name', 
                   how='left', indicator=True)
df_all
```
