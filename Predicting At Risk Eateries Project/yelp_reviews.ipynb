# Get all location names

```python
url_main = 'https://healthspace.com/Clients/Ohio/Butler/Web.nsf/Food-CityList?OpenView&Count=999&'
```

```python
r_main = requests.get(url_main)
soup = BeautifulSoup(r_main.text, 'html.parser')
a = soup.find_all('a') # get all a tags
location_names = [a[i].get_text().replace(' ', '+') for i in range(3, len(a))] # location appear after first 3
```

```python
location_names
```

```python
# all pages to be considered
pages = [*range(0, 200, 10)]
```

```python
# main yelp url
yelp_urls = 'https://www.yelp.com/search?find_desc=&find_loc={}%2C+OH&start={}'
```

```python
all_loc_w_page = []
```

```python
def get_location_url(loc, page):
    return yelp_urls.format(loc,page)
```

```python
for location in location_names:
    for page in pages:
        all_loc_w_page.append(get_location_url(location, page))
```

## Run Scraping Code

```python
pages = [*range(0, 100, 10)] #for looping through pages of reviews
pages
```

```python
list_links = all_loc_w_page[260:]
```

```python
for index,i in enumerate(list_links):
    if 'oxford' in i.lower():
        list_links.pop(index)
```

```python
list_links
```

```python
url_domain = 'https://www.yelp.com/'
```

```python
# initial lists to store result
review_list = []
name_for_append = []
address_list = []
phone_list = []
date_of_review = []
```

```python
for page in list_links:
    print(page)
    r_main = requests.get(page)
    soup = BeautifulSoup(r_main.text, 'html.parser')
    tds = soup.find_all('a', {'class': 'css-1m051bw'})
    for td in tds:
        link = td['href']
        link_request = ''.join([url_domain, link])
        link_request += '?osq=Restaurants&start={}'
        for p in pages:
            l = link_request.format(p)
            print(l)
            if '//biz/' in l:
                r = requests.get(l)
                if r.text.find('You may need permission to access this page') != -1:
                    break
                restaurant = BeautifulSoup(r.text, 'html.parser')
                name = restaurant.find('h1',{'class':'css-1se8maq'})
                if name is None:
                    name = 'NA'
                else:
                    name = name.contents[0]
                print(name)

        #             address = restaurant.find('p',{'class':'css-qyp8bo'}).contents[0]
                address = restaurant.find('p',{'class':'css-qyp8bo'})
                if address is None:
                    address = 'NA'
                else:
                    address = address.contents[0]

                review = []
                comments = restaurant.find_all('li', {'class': 'margin-b5__09f24__pTvws border-color--default__09f24__NPAKY'})

                    # Loop through comments
                for comment in comments:
                    try:
                        date = comment.find('span', {'class': 'css-chan6m'}).get_text()
                        text = comment.find('span', {'class': 'raw__09f24__T4Ezm', 'lang': 'en'}).get_text()
                    except:
                        date = 'NA'
                        text = 'NA'
                    review.append(text)
                    review_list.append(text)
                    date_of_review.append(date)

                ## PHONE NUMBER
                divs = restaurant.find_all('div', {'class': 'css-1vhakgw border--top__09f24__exYYb border-color--default__09f24__NPAKY'})
                for div in divs:
                    p_name = div.find('p', {'class': 'css-na3oda'})
                #     print(p_name)
                    try:
                        if p_name.get_text() == 'Phone number':
                            try:
                                p_phone = div.find('p', {'class': 'css-1p9ibgf', 'data-font-weight': 'semibold'}).get_text()
                            except:
                                p_phone = 'NA'
                    except:
                        p_phone = 'NA'

                #append other info    
                for i in range(0,len(review)): 
                    name_for_append.append(name)
                    address_list.append(address)
                    phone_list.append(p_phone)
```
